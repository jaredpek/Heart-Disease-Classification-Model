## IMPORT TOOLS AND LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## IMPORT DATA

```python
heartDiseaseData = pd.read_csv('heart-disease.csv')
heartDisease = pd.DataFrame(heartDiseaseData)
```

## VISUALISE DATA

##### RELATIONSHIP BETWEEN SEX AND HEART DISEASE
TYPES OF SEX:
- SEX 0 = FEMALE
- SEX 1 = MALE

TYPES OF TARGETS:
- TARGET 0 = NO HEART DISEASE
- TARGET 1 = HEART DISEASE

```python
sex_target_fig, sex_target = plt.subplots(figsize=(13, 7))

sex = ['Male', 'Female']
sex_axis = np.arange(len(sex))

female_heartDisease = len(heartDisease[(heartDisease['sex'] == 0) & (heartDisease['target'] == 1)])
female_noHeartDisease = len(heartDisease[(heartDisease['sex'] == 0) & (heartDisease['target'] == 0)])
male_heartDisease = len(heartDisease[(heartDisease['sex'] == 1) & (heartDisease['target'] == 1)])
male_noHeartDisease = len(heartDisease[(heartDisease['sex'] == 1) & (heartDisease['target'] == 0)])

sex_heartDisease = [male_heartDisease, female_heartDisease]
sex_noHeartDisease = [male_noHeartDisease, female_noHeartDisease]

sex_target.bar(sex_axis - 0.2, sex_heartDisease, 0.4, label='Heart Disease')
sex_target.bar(sex_axis + 0.2, sex_noHeartDisease, 0.4, label='No Heart Disease')
sex_target.set(title='Relationship between Gender and Heart Disease', xlabel='Gender', ylabel='Number of People', xticks=sex_axis, xticklabels=sex)
sex_target.legend(title='Legend', frameon=True);
```

##### RELATIONSHIP BETWEEN CP AND HEART DISEASE
TYPES OF CHEST PAIN:
- CP 0 = TYPICAL ANGINA
- CP 1 = ATYPICAL ANGINA
- CP 2 = NON-ANGINAL PAIN
- CP 3 = ASYMPTOMATIC

TYPES OF TARGET:
- TARGET 0 = NO HEART DISEASE
- TARGET 1 = HEART DISEASE

```python
cp_target_fig, cp_target = plt.subplots(figsize=(13, 7))

painType = ['Typical Angina', 'Atypical Angina', 'Non-Anignal Pain', 'Asymptomatic']
painType_axis = np.arange(len(painType))

pain0_heartDisease = len(heartDisease[(heartDisease['cp'] == 0) & (heartDisease['target'] == 1)])
pain0_noHeartDisease = len(heartDisease[(heartDisease['cp'] == 0) & (heartDisease['target'] == 0)])
pain1_heartDisease = len(heartDisease[(heartDisease['cp'] == 1) & (heartDisease['target'] == 1)])
pain1_noHeartDisease = len(heartDisease[(heartDisease['cp'] == 1) & (heartDisease['target'] == 0)])
pain2_heartDisease = len(heartDisease[(heartDisease['cp'] == 2) & (heartDisease['target'] == 1)])
pain2_noHeartDisease = len(heartDisease[(heartDisease['cp'] == 2) & (heartDisease['target'] == 0)])
pain3_heartDisease = len(heartDisease[(heartDisease['cp'] == 3) & (heartDisease['target'] == 1)])
pain3_noHeartDisease = len(heartDisease[(heartDisease['cp'] == 3) & (heartDisease['target'] == 0)])

painType_heartDisease = [pain0_heartDisease, pain1_heartDisease, pain2_heartDisease, pain3_heartDisease]
painType_noHeartDisease = [pain0_noHeartDisease, pain1_noHeartDisease, pain2_noHeartDisease, pain3_noHeartDisease]

cp_target.bar(painType_axis - 0.2, painType_heartDisease, 0.4, label='Heart Disease')
cp_target.bar(painType_axis + 0.2, painType_noHeartDisease, 0.4, label='No Heart Disease')
cp_target.set(title='Relationship between Chest Pain and Heart Disease', xlabel='Type of Chest Pain', ylabel='Number of People', xticks=painType_axis, xticklabels=painType)
cp_target.legend(title='Legend', frameon=True);
```

##### AGE DISTRIBUTION OF PATIENTS

```python
age_fig, age_hist = plt.subplots(figsize=(13, 7))

age_hist.hist(heartDisease['age'], edgecolor='black')
age_hist.set(title='Age Distribution', xlabel='Age', ylabel='Frequency')
age_hist.set_ylim(0, 70);
```

##### RELATIONSHIP BETWEEN THALACH AND AGE
- THALACH = MAXIMUM HEART RATE

TYPES OF TARGETS:
- TARGET 0 = NO HEART DISEASE
- TARGET 1 = HEART DISEASE


```python
thalach_age__fig, thalach_age = plt.subplots(figsize=(13, 7))

thalach_age.scatter(heartDisease[heartDisease['target'] == 0]['age'], heartDisease[heartDisease['target'] == 0]['thalach'], label='No Heart Disease', color='green')
thalach_age.scatter(heartDisease[heartDisease['target'] == 1]['age'], heartDisease[heartDisease['target'] == 1]['thalach'], label='Heart Disease', color='red')
thalach_age.set(title='Relationship between Maximum Heart Rate and Age', xlabel='Age', ylabel='Maximum Heart Rate')
thalach_age.legend(title='Legend', frameon=True)
thalach_age.set_xlim(20, 90)
thalach_age.set_ylim(60, 240);
```

##### DATA CORRELATION

```python
hD_corrMatrix = heartDisease.corr()
corrMatrix_fig, corrMatrix_ax = plt.subplots(figsize=(20, 13))
sns.heatmap(hD_corrMatrix, annot=True, fmt='.2f', ax=corrMatrix_ax)
corrMatrix_ax.set(title='Data Correlation');
```
